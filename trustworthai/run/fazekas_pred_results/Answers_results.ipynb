```python
import os
from natsort import natsorted
```

```python
files = os.listdir(".")
```

```python
files = natsorted([f for f in files if ".txt" in f])
```

```python
for file in files:
    with open(file, 'r') as f:
        lines = f.readlines()
        
        print(file)
        print("-------------------------")
        for line in lines:
            if "for model" in line:
                print(line)
                
        print("-------------------------\n")
```
